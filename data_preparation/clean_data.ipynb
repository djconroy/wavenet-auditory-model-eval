```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
DATASET = "timit/core-test-set"
```

```python
parent_dir = Path.cwd().parent
comparison_metrics_dir = Path(parent_dir, "comparison_metrics", DATASET)
phonemes_file = Path(comparison_metrics_dir, "phonemes.csv")
cleaned_phonemes_file = Path(comparison_metrics_dir, "cleaned_phonemes.csv")
```

### Load phonemes.csv into a pandas DataFrame
Some entries for the SNR column are NA, which is informative and not indicative of missing data, so they shouldn't be considered as indicative of missing data by pandas. Only entries in phonemes.csv with the value nan or 0.0 will be considered as n/a values. Note: No entries for the SPL or SNR columns have the value 0.

```python
phonemes_df = pd.read_csv(phonemes_file, na_values="nan", keep_default_na=False)
```

### Set entries with the value 0.0 to np.nan to make all n/a entries consistent

```python
phonemes_df = phonemes_df.replace(0.0, np.nan)
```

```python
phonemes_df[phonemes_df == 0.0][23:39]
```

```python
phonemes_df[23:39]
```

### Verify entries with 0.0 now have the value nan instead

```python
phonemes_df[phonemes_df == 0].count().sum()
```

```python
phonemes_df.isna().sum().sum()
```

### Show the numbers of entries for each kind of phoneme

```python
phonemes_df["Phoneme"].value_counts()
```

### Remove entries with epi or pau

The values epi and pau respectively correspond to epenthetic silence and a pause. Rows with them will be removed.

```python
phonemes_df = phonemes_df[~phonemes_df["Phoneme"].isin(["epi", "pau"])]
```

```python
phonemes_df["Phoneme"].value_counts()
```

### Save the cleaned data in cleaned_phonemes.csv

```python
phonemes_df.to_csv(cleaned_phonemes_file, na_rep="nan", index=False, lineterminator="\n")
```
